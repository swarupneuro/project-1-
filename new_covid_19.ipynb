```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score # Import all necessary metrics
from sklearn.impute import SimpleImputer

# Load the dataset
df = pd.read_csv("/content/covid 19 csv.csv")

# Clean column names by stripping leading/trailing spaces
df.columns = df.columns.str.strip()

# Replace '-' with NaN in all columns and convert to numeric
for col in df.columns:
    if col not in ['State/UnionTerritory', 'High_Confirmed']: # Adjust excluded columns as necessary
        df[col] = pd.to_numeric(df[col], errors='coerce')

# Create a binary target: 1 if Confirmed > average, else 0
if 'Confirmed' in df.columns:
    # Ensure 'Confirmed' column is numeric before calculating mean
    df['Confirmed'] = pd.to_numeric(df['Confirmed'], errors='coerce')
    avg_confirmed = df['Confirmed'].mean()
    df['High_Confirmed'] = (df['Confirmed'] > avg_confirmed).astype(int)
else:
    raise ValueError("Column 'Confirmed' not found. Please check column names.")

# One-hot encode the 'State/UnionTerritory' column
if 'State/UnionTerritory' in df.columns:
    df = pd.get_dummies(df, columns=['State/UnionTerritory'], prefix='State')
else:
    raise ValueError("Column 'State/UnionTerritory' not found. Please check column names.")

# Drop irrelevant columns
cols_to_drop = ['Sno', 'Date', 'Time', 'Confirmed']
existing_cols_to_drop = [col for col in cols_to_drop if col in df.columns]
df = df.drop(existing_cols_to_drop, axis=1)

# Separate features and target
X = df.drop('High_Confirmed', axis=1)
y = df['High_Confirmed']

# Impute missing values in the feature set X
imputer = SimpleImputer(strategy='mean')
X_imputed = imputer.fit_transform(X)

# Convert imputed array back to DataFrame
X_imputed = pd.DataFrame(X_imputed, columns=X.columns, index=X.index)

# Train-test split using the imputed data
# Using test_size=1/3 as per your ipython-input-19 cell
X_train, X_test, y_train, y_test = train_test_split(X_imputed, y, test_size=1/3, random_state=0)

# Train the logistic regression model
Model1 = LogisticRegression(max_iter=1000) # Added max_iter to avoid potential convergence warnings
Model1.fit(X_train, y_train)

# Predicting the test set results
prediction1 = Model1.predict(X_test)

# Evaluate the model using confusion matrix and accuracy
print("Confusion Matrix:\n", confusion_matrix(y_test, prediction1))
print("Accuracy:", accuracy_score(y_test, prediction1))
print("Classification Report:\n", classification_report(y_test, prediction1)) # Added classification report for more detailed evaluation

# Optional: visualize correlation heatmap
plt.figure(figsize=(12,8))
# Use numeric_only=True to avoid errors with non-numeric types
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```
cm = confusion_matrix(y_true = y_test, y_pred = prediction1)
#plot_confusion_matrix(cm,level,title = "confusion_matrix")
sns.heatmap(cm, annot=True, cmap="Blues", fmt="d")
plt.xlabel("Predicted Labels")
plt.ylabel("True Labels")
plt.title("Confusion Matrix")
plt.show()
```
